```python
sm.qqplot(LabTAT["Laboratory 1"], line = 'q')
plt.title('Laboratory 1')
sm.qqplot(LabTAT["Laboratory 2"], line = 'q')
plt.title('Laboratory 2')
sm.qqplot(LabTAT["Laboratory 3"], line = 'q')
plt.title('Laboratory 3')
sm.qqplot(LabTAT["Laboratory 4"], line = 'q')
plt.title('Laboratory 4')
plt.show()
```

```python
test_statistic , p_value = stats.f_oneway(LabTAT.iloc[:,0],LabTAT.iloc[:,1],LabTAT.iloc[:,2],LabTAT.iloc[:,3])
print('p_value =',p_value)
```

# Step 5
Compare P_value with singificance value i.e. 0.05
if p > alpha value, fail to reject null hypothesis
if p < alpha value, reject null hypothesis and accept alternate hypothesis

```python
alpha = 0.05
if p_value >= alpha:
    print('p_Value=', p_value, 'alpha=', alpha)
    print("Accept null hypothesis\nthere is no significant difference in average TAT among the different laboratories.")
else:
    print('p_Value=', p_value, 'alpha=', alpha)
    print("Reject null hypothesis\nthere is significant difference in average TAT among the different laboratories.")
```
